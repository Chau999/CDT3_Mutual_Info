# Independence test
Before we quantify the amount of dependency exists between variables, we have to make sure such dependencies exist in the first hand. This falls into the problem of hypothesis testing.

```python
def kernel_independent_test(x, y, sig_lvl=0.05, verbose=True):
    
    x = x.reshape(len(x), 1)
    y = y.reshape(len(y), 1)
    
    num_samples = len(x)
    
    kernelX=GaussianKernel()
    kernelY=GaussianKernel()

    #example usage of HSIC spectral test with random Fourier feature approximation
    myspectralobject = HSICSpectralTestObject(num_samples, kernelX=kernelX, kernelY=kernelY, 
                                              kernelX_use_median=True, kernelY_use_median=True,
                                              rff=True, num_rfx=20, num_rfy=20, num_nullsims=1000)

    pvalue = myspectralobject.compute_pvalue(x, y)

    if verbose:
        print ("Spectral test p-value:", pvalue)

        if pvalue < sig_lvl:
            print("Reject null --> They are dependent")
        else:
            print("Not reject null --> they are independent")
    
    return pvalue
```

```python
# Write functions to produce data for tests

def gen_data(xrange=[-10, 10], num_points=10000, noise_lvl = 3, fun_struc="Linear", error_type="Gaussian"):
    x = np.linspace(xrange[0], xrange[1], num_points)
    
    if error_type == "Gaussian":
        error = np.random.normal(0, 1, num_points)
    elif error_type == "Exponential":
        error = np.random.exponential(scale=1, size=num_points)
    
    if fun_struc == "Linear":
        y = x + noise_lvl * error
    elif fun_struc == "Quadratic":
        y = 0.5* x ** 2 + noise_lvl * error
    elif fun_struc == "Exponential":
        y = np.exp(x) + noise_lvl * error
    elif fun_struc == "Sine":
        y = np.sin(x) + noise_lvl * error
    elif fun_struc == "Independent":
        y = np.random.choice(x, size=num_points, replace=False) + noise_lvl * error
    elif fun_struc == "SineX":
        y = np.sin(x) * x + noise_lvl * error
    elif fun_struc == "Sine/X":
        y = np.sin(x) / x + noise_lvl * error
    
    return x, y

def plot_data(x, y):
    plt.plot(x, y)
    plt.show()
```

```python
def MINE(x, y, n_epoch=200, representation='DV', hidden_parameters=10, 
         optimizer='Adam', batch_size = 32, return_loss_plot=False):
    
    class Additive_Net(nn.Module):  # possible choices for the net are possible
    # this has 5*H + 1 parameters
        def __init__(self, H=10):
            super(Additive_Net, self).__init__()
            self.fc1 = nn.Linear(1, H)
            self.fc2 = nn.Linear(1, H)
            self.fc3 = nn.Linear(H, 1)

        def forward(self, x, y):
            h1 = F.relu(self.fc1(x)+self.fc2(y))
            h2 = self.fc3(h1)
            return h2    

    model = Additive_Net(hidden_parameters)    
    
    if (optimizer=='Adam'): 
        optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
    elif (optimizer=='SGD'): 
        optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

    plot_loss = []
    
    # convert to torch variable
    x_sample = Variable(torch.from_numpy(x).type(torch.FloatTensor), requires_grad = True)
    y_sample = Variable(torch.from_numpy(y).type(torch.FloatTensor), requires_grad = True)
    
    splits=np.arange(0, len(x)+batch_size, batch_size)  # find the indeces splitting the data in mini-batches
    
    for epoch in tqdm(range(n_epoch)):
                
        y_shuffle=np.random.permutation(y)  # shuffle to be able to estimate independent average
        y_shuffle = Variable(torch.from_numpy(y_shuffle).type(torch.FloatTensor), requires_grad = True)    

        epoch_loss = 0
        
        for batch in range(len(splits)-1):

            x_batch = x_sample[splits[batch]:splits[batch+1]]
            y_batch = y_sample[splits[batch]:splits[batch+1]]
            y_shuffle_batch = y_shuffle[splits[batch]:splits[batch+1]]
            
            pred_xy = model(x_batch, y_batch)  # empirical average using joint prob.
            pred_x_y = model(x_batch, y_shuffle_batch)  # empirical average using shuffled (independent hypothesis) prob.

            if (representation=='DV'):
                ret = torch.mean(pred_xy) - torch.log(torch.mean(torch.exp(pred_x_y)))
            elif (representation=='f'):
                ret = torch.mean(pred_xy) - torch.mean(torch.exp(pred_x_y) - 1)

            
            loss = - ret  # maximize
            epoch_loss += loss
            model.zero_grad()
            loss.backward()
            optimizer.step()
        
        plot_loss.append(epoch_loss.data.numpy()/(len(splits)-1))  # return the loss averaged over batches
        
    # to compute the average estimation I discard the first 25% of steps (arbitrary choice)
    if (return_loss_plot):
        return -np.mean(plot_loss[np.int(n_epoch/4):]), np.var(plot_loss[np.int(n_epoch/4):]), plot_loss
    else: 
        return -np.mean(plot_loss[np.int(n_epoch/4):]), np.var(plot_loss[np.int(n_epoch/4):])
    
```

```python
f, ax = plt.subplots(3, 3, sharey=False, figsize = (21, 18))
counter = 0
for fun in ["Linear", "Quadratic", "Independent"]:    
    kernel_p_value = []
    mic_value = []
    mi_value = []
    noise_range = np.linspace(0, 1000, 20)
    for noise in noise_range:
        x, y = gen_data(fun_struc=fun, noise_lvl=noise)
        
        # Kernel Independent Test
        kernel_p_value.append(kernel_independent_test(x, y, verbose=False))

        # MIC
        mine = mpy.MINE(alpha=0.5, est="mic_approx")
        mine.compute_score(x, y)
        mic_value.append(mine.mic())
        
        # MI
        mi_value.append(mutual_info_regression(x.reshape(-1, 1),y.reshape(-1, )))

    ax[0,counter].plot(noise_range ,kernel_p_value) 
    ax[0,counter].axhline(0.05,linewidth=4, color='r')
    ax[0,counter].set_xlabel("Noise_level")
    ax[0,counter].set_ylabel("p-values")
    ax[0,counter].set_title("Functional_structure: %s (KI)" %fun)
    
    ax[1,counter].plot(noise_range, mic_value) 
    ax[1,counter].set_xlabel("Noise_level")
    ax[1,counter].set_ylabel("MIC_Values")
    ax[1,counter].set_title("Functional_structure: %s (MIC)" %fun)

    ax[2,counter].plot(noise_range, mi_value) 
    ax[2,counter].set_xlabel("Noise_level")
    ax[2,counter].set_ylabel("Mutual_Information_Values")
    ax[2,counter].set_title("Functional_structure: %s (Mutual_Information)" %fun)
    
    counter += 1

plt.show()
```

```python
def plot_with_scores_test(x, y, scatter=True):
    mine.compute_score(x, y)
    textstr = '\n'.join((
    "KI p-value - %.2f" % kernel_independent_test(x, y, verbose=False),
    "MIC value - %.2f" % mine.mic(),
    "MI value - %.2f" % mutual_info_regression(x.reshape(-1, 1),y.reshape(-1, ))))    
    props = dict(boxstyle='round', facecolor='white', alpha=0.5)
    f, ax = plt.subplots(figsize=(8,5))
    if scatter:
        ax.scatter(x, y)
    else:
        ax.plot(x, y)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.text(0.5, 0.85, textstr, transform=ax.transAxes, fontsize=16,bbox=props)
```
